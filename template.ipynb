```python
import pygame
import random
```

# Setup Game

## Window Size

```python
WIDTH = 360
HEIGH = 360
FPS = 30
```

## init Pygame

```python
pygame.init()  #init pygame and create window
screen = pygame.display.set_mode((WIDTH,HEIGH)) #ekran gsterecek 
pygame.display.set_caption("KODLUYORUZ_RL") #oyunun isminin yazd ksm
clock = pygame.time.Clock()
```

# Game Loop

```python
running = True
while running:
    # keep loop running at the right speed
    clock.tick(FPS)
    
    #process input
    for event in pygame.event.get(): # pygame eventlerini getirdi
        if event.type == pygame.QUIT: # event QUIT ise
            running = False           # while looptan k
            
    #update
    
    # draw / render(s)how)
    screen.fill(green) #RGB (0,255,0)
    
    
    # after drawing flip dispalay
    pygame.display.flip()
    
pygame.quit()                         # looptan knca oyunu kapat
```
